```python
import random

def estimate_pi(num_points):
    inside_circle = 0

    for i in range(num_points):
        x = random.uniform(-1, 1)  # Randomly generate x coordinate between -1 and 1
        y = random.uniform(-1, 1)  # Randomly generate y coordinate between -1 and 1

        if x**2 + y**2 <= 1:
            inside_circle += 1

        if (i+1) % 1_000_000 == 0:  # Every 1 million rounds
            pi_estimate = 4 * inside_circle / (i+1)
            print(f"After {i+1} rounds, the estimated value of pi is: {pi_estimate}")

    # Return the final estimated value of pi
    return 4 * inside_circle / num_points

# Number of points to sample
num_points = 100_000_000  # 100 million points

# Call the function and print the final estimated value of pi
final_pi_estimate = estimate_pi(num_points)
print("The estimated value of pi after 100 million rounds:", final_pi_estimate)
```
